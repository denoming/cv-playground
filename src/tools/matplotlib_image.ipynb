```python
import matplotlib.pylab as plt
import numpy as np
from PIL import Image
from common import CV_DATA_DIR

%matplotlib inline
```

```python
image_path = CV_DATA_DIR / 'misc' / 'stinkbug.png'
image = np.asarray(Image.open(image_path))
print(f'Shape of the image: {image.shape}')
```

```python
imgplot = plt.imshow(image)
```

```python
lum_image = image[:, :, 0]
print(f'Shape of the image: {lum_image.shape}')
```

```python
plt.figure()
plt.subplot(131)
plt.imshow(lum_image)
plt.colorbar()
plt.subplot(133)
plt.hist(lum_image.ravel(), bins=range(256), fc='k', ec='k')
plt.suptitle('Image and Histogram')
plt.show()
```

```python
plt.imshow(lum_image, cmap="hot")
plt.colorbar()
```
