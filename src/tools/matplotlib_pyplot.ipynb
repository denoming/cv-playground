```python
import matplotlib.pylab as plt
import numpy as np
```

```python
# Plot y values by default (if one sequence is provided)
plt.plot([1, 2, 3, 4, 5])
plt.ylabel('y-values')
plt.show()
```

```python
# Plot x and y values
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.show()
```

```python
# Plot numpy array
t1 = np.arange(0., 5., 0.2)
plt.plot(t1, t1, 'r--', t1, t1**2, 'rs', t1, t1**3, 'g^')
plt.show()
```

```python
data = {
    'a': np.arange(50),
    'b': np.arange(50) + 10 * np.random.randn(50),
    'c': np.random.randint(0, 50, 50),
    'd': np.abs(np.random.randn(50)) * 100
}

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('a')
plt.ylabel('b')
plt.show()
```

```python
names = ['groupA', 'groupB', 'groupC']
values = [1, 10, 100]

plt.figure(figsize=(9,3)) # Set figure size
# Make a bar plot
plt.subplot(131)
plt.bar(names, values, color='b')
# Make a scatter plot y vs x
plt.subplot(132)
plt.scatter(names, values, c='g')
# Make a plot x as lines
plt.subplot(133)
plt.plot(names, values, 'r-', linewidth=3.0)
# Show all figures
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
# Manipulate lines properties
x1 = np.array([1, 2, 3, 4])
x2 = np.square(x1)
y1 = np.array([1, 4, 9, 16])
y2 = np.sqrt([1, 4, 9, 16])

plt.figure(num='Figure 1')
lines = plt.plot(x1, y1, x2, y2)
plt.setp(lines, color='g', linewidth=1.0, marker='o')
plt.show()
```

```python
# Working with multiple figures and axes
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(num='Figure 2')

# Subplot #1
plt.subplot(211) # 2 - numrows, 1 - numcols, 1 - plot number from 1 to numrows*numcols
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.grid(True)
plt.text(1.0, 0.5, r'$\mu=100,\ \sigma=15$') # put text at data coordinate location using math expr

# Subplot #2
plt.subplot(2, 1, 2)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.annotate('max', xy=(1.0, 1.0), xytext=(1.3, 0.7),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.xlabel('time')
plt.ylabel('amplitude')
plt.grid(True)

plt.show()
```

```python
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

plt.plot(x, y, 'r')
plt.yscale('log')
plt.title('log')
plt.grid(True)
plt.show()
```
